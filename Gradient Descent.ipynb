```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(47)
N = 150
X = np.random.rand(N)
X = np.sort(X)
# y_pred = ax+b
a = 1.5
b = 2.5
sigma = 0.2
#y_true
y = a*X + b + np.random.randn(N)*sigma

plt.plot(X,y, "b-o")
plt.plot(X, a*X+b, "r-o")
```
